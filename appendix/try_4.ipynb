```python
import pandas as pd


df = pd.read_csv("Data-able.csv")
```

```python
df.head()
```

```python
# Identify columns with 'datetime' in their name
datetime_columns = [col for col in df.columns if 'datetime' in col.lower()]

print("Columns with 'datetime' in their name:", datetime_columns)

# Drop those columns
df = df.drop(columns=datetime_columns)

print("\nDataFrame after removing columns with 'datetime':")
print(df.head())

# Optional: Save the cleaned DataFrame
df.to_csv('cleaned_file.csv', index=False)
```

```python
#missing data??

import pandas as pd


# 1. Identify columns with numerical data
numerical_columns = df.select_dtypes(include=['number']).columns.tolist()

print("Numerical columns detected:", len(numerical_columns))

# 2. Check for rows with missing data in numerical columns
missing_data = df[numerical_columns].isnull()

# 3. Summarize missing data
missing_summary = missing_data.sum()  # Count of missing values per column
print("\nMissing values summary per column:")
print(missing_summary)

# 4. Investigate rows with missing data
rows_with_missing = df[missing_data.any(axis=1)]  # Rows where any numerical column is missing

print("\nRows with missing numerical data:")
print(rows_with_missing)

# Optional: Save rows with missing data for manual review
rows_with_missing.to_csv("rows_with_missing_data.csv", index=False)
```

```python
#fill all missing with the median:

# Fill missing numerical values with the median
numerical_columns = df.select_dtypes(include=['number']).columns

# Replace missing values in each numerical column with its median
df[numerical_columns] = df[numerical_columns].apply(lambda col: col.fillna(col.median()))

# Check if there are still any missing values
print("\nMissing values summary after filling with median:")
print(df[numerical_columns].isnull().sum())
```

```python
# Drop rows with missing values
df_cleaned = df.dropna()

# Check the new shape of the DataFrame
print("Shape after dropping rows with missing values:", df_cleaned.shape)

# Save the cleaned DataFrame (optional)
df_cleaned.to_csv('cleaned_data_for_eda.csv', index=False)
```
