```python
import requests
import pandas as pd



# Function to make API call for each location
def get_weather_data(location, date):
    api_key = 'WTCKQMFXBWJSST6TGTZCVZF5Y'
    url = f'https://weather.visualcrossing.com/VisualCrossingWebServices/rest/services/timeline/{location}/{date}?key={api_key}'
    
    # Make the API request
    response = requests.get(url)
    
    if response.status_code == 200:
        data = response.json()
        # Extract the full day's weather information
        days_info = data.get('days', [{}])[0]  # Extract the first day's data
        hours_info = days_info.get('hours', [])  # Extract hourly weather data
        
        return days_info, hours_info
    else:
        print(f"API call failed for {location} on {date} with status code {response.status_code}")
        return None, None

# Loop through the first 3 rows to request weather data
for index, row in df_money.iterrows():
    latitude = row['latitude']
    longitude = row['longitude']
    date = row['date']
    
    if (latitude is None) or (longitude is None):
        continue
    
    # Format location with 6 decimal places
    location = f"{latitude:.6f},{longitude:.6f}"
    
    # Request weather data
    daily_info, hourly_info = get_weather_data(location, date)
    
    if daily_info:
        # Add the daily data to the dataframe
        for key, value in daily_info.items():
            # Ensure we're setting a single value, not a list of values
            if isinstance(value, list):  # For fields that might be lists, such as 'hours'
                df_money.loc[index, f'weather_{key}'] = ', '.join(map(str, value))  # Store list as string
            else:
                df_money.loc[index, f'weather_{key}'] = value
        
        # Add hourly data as well, if available
        for hour_index, hour in enumerate(hourly_info):
            for key, value in hour.items():
                # Ensure we're setting a single value, not a list of values
                if isinstance(value, list):
                    df_money.loc[index, f'weather_hour_{hour_index}_{key}'] = ', '.join(map(str, value))
                else:
                    df_money.loc[index, f'weather_hour_{hour_index}_{key}'] = value

# Now drop the 'weather_hours' column if it exists
        if 'weather_hours' in df_money.columns:
            df_money.drop(columns=['weather_hours'], inplace=True)
            
            
# Print the DataFrame with the full weather data
print(df_money)
```

```python
# Replace all 0s in the dataset with the column mean
def replace_zeros_with_mean(df):
    for column in df.select_dtypes(include=['number']).columns:  # Apply only to numeric columns
        column_mean = df[column][df[column] != 0].mean()  # Calculate mean excluding zeros
        df[column] = df[column].replace(0, column_mean)  # Replace zeros with the mean
    return df

# Apply the function
new_df = replace_zeros_with_mean(new_df)
```
