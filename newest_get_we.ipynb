```python
import pandas as pd
import json

# Function to clean and process the 'Weather' column
def process_weather_column(df, weather_column):
    def fix_and_parse_json(json_string):
        try:
            # Replace single quotes with double quotes
            #fixed_string = json_string.replace("'", '"')
            # Convert string to dictionary
            return json.loads(json_string)
        except json.JSONDecodeError:
            print("we faited to parse")
            return None  # Return None if parsing fails
    
    # Fix and parse the JSON column
    parsed_weather = df[weather_column].dropna().apply(fix_and_parse_json)
    
    # Drop rows where parsing failed
    parsed_weather = parsed_weather.dropna()
    
    # Expand JSON into separate columns
    weather_expanded = pd.json_normalize(parsed_weather)
    
    # Combine expanded columns with original dataframe
    df = df.drop(columns=[weather_column])  # Remove original Weather column
    df = pd.concat([df, weather_expanded], axis=1)
    
    return df


# Assuming 'Weather' is the column with JSON data
df_processed = process_weather_column(df_done, 'Weather')

# Save the processed dataframe to a new file
df_processed.to_csv('processed_weather_data.csv', index=False)
```

```python
import pandas as pd
import json

# Load your dataframe
df_done = pd.read_csv('differentformat.csv')

# Define a function to parse the weather column
def parse_weather(weather_str):
    try:
        # Convert the string to a dictionary
        weather_dict = json.loads(weather_str.replace("'", '"'))  # Ensure valid JSON format
        return pd.Series({
            'temperature': weather_dict.get('temp'),
            'precipitation': weather_dict.get('precip'),
            'humidity': weather_dict.get('humidity'),
            'weather_description': weather_dict.get('conditions'),
        })
    except (json.JSONDecodeError, TypeError):
        # Return NaN for invalid or missing entries
        return pd.Series({'temperature': None, 'precipitation': None, 'humidity': None, 'weather_description': None})

# Apply the parsing function to the weather column
parsed_weather = df_done['Weather'].apply(parse_weather)

# Merge the parsed columns back into the original dataframe
df_done = pd.concat([df_done, parsed_weather], axis=1)

# Save the cleaned dataframe for review
df_done.to_csv('parsed_weather_data.csv', index=False)

print("Weather data has been parsed and saved as 'parsed_weather_data.csv'.")
```
